```python
### General Function to calculate return of compound interest 
def compoundInterest(p, r, t):
    A =  p * (1+r)**t
    return A

p = 100
r = 0.05
t = 5
print('The return value after 5 ys with compounding interest is: {}'.format(compoundInterest(p, r, t)))
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

t = list(range(26))
p = 100
r = 0.01
returns = [compoundInterest(p, r, t_) for t_ in t]
plt.plot(t, returns)
plt.xlabel('Year')
plt.ylabel('Return Value')
plt.title('Compound Interest')
plt.plot()
```

```python
# Recall slicing 
print(returns[1:])
print(returns[:-1])
```

```python
# Calculate interest earning === pct_change
interest_earn = [i - j for i, j in zip(returns[1: ], returns[:-1])]
plt.plot(list(range(1, 26)), interest_earn)
plt.xlabel('Year')
plt.ylabel('Interest Earn')
plt.plot()
```

```python
rate = [i/j for i,j in zip(interest_earn,returns[:-1])]
print(rate)
```

### Changing r 

Sometime r is not going to be the same every year. how this might impact our interest.

```python
#How can we make a list of the rate 10% in 10 years, then 2% in the 15 years after?
#Multiplying a list in python will make it expand to have that many identical elements
rates = [.1]*10+[.02]*15
print(rates)
```

we can see how for 10 years we earn 10% interest, but then we earn 2% interest during the last 15 years. Creating a compounding equation and running it through our results are as follows.

```python
def compound(P, rates):
    returns  = [P]
    for rate in rates: 
        returns.append(returns[-1] * (1 + rate))
    return returns
```

```python
compound(100, rates)
```

```python
plt.plot(list(range(0,26)), compound(100, rates))
```

Let's get data for two sets of rates. One the same as before, and then a second where the rates are reversed.

```python
rates = [[.1]*10+[.02]*15, [.02]*15+[.1]*10]
```

```python
rates
```

```python
#Let's modify the function to take multiple sets of rates and return a nested list of the values
def compound(P,rates):
    data = []
    for rates_i in rates:
        A = [P]
        for rate in rates_i:
            value = A[-1]*(rate+1)
            A.append(value)
        data.append(A)
    return data
print(compound(100,rates))
```

```python
import matplotlib.pyplot as plt
def plotCompoundInterest(P,rates,names):
    #Add a name input to hold our investment choice names
    data = compound(P,rates)
    for A,name in zip(data,names):
        t = list(range(len(A)))
        plt.plot(t,A,label=name)
    plt.xlabel("t")
    plt.ylabel("A")
    plt.title("Compound Interest")
    plt.legend()
    plt.show()
plotCompoundInterest(100,rates,["Choice A","Choice B"])
```

### Interest with Withdrawals

The analysis we are doing is great when we want to see things like the value of retirement contributions, but in the real world we may take money out of our investments. Let's define a principal amount, a yearly withdrawl amount, and a list of lists holding rates for three different rate scenarios.


```python
#Now, let's say we have a trust fund of 1 million dollars
#And we want to see what happens as we take out an amount at the beginning of every year: $70,000
P = 1000000
rates = [[.02]*25,[.04]*25,[.06]*25]
withdraw = 70000
```

```python
def compoundWithdraw(P,rates,withdraw):
    data = []
    for rate_i in rates:
        A = [P]
        for rate in rate_i:
            value = A[-1]*(rate+1)
            value = max(value-withdraw,0)
            #Withdraw at the beginning of the period, after we have gotten our yearly interest payment
            #Set the max of 0 and this number since we can't go into debt
            A.append(value)
        data.append(A)
    return data
print(compoundWithdraw(P,rates,withdraw))
```

We get an increase in our value based on the rate but withdraw an amount afterwards every period. We need to use the max function to make sure that our value does not go below $0.

Time to plot.

```python
def plotCompoundWithdraw(P,rates,names,withdraw):
    data = compoundWithdraw(P,rates,withdraw)
    for A,name in zip(data,names):
        t = list(range(len(A)))
        plt.plot(t,A,label=name)
    plt.xlabel("t")
    plt.ylabel("A")
    plt.title("Compound Interest with Withdraws")
    plt.legend()
    plt.show()
plotCompoundWithdraw(P,rates,["r=.02","r=.04","r=.06"],withdraw)
```

As we have seen before, we can use a function formatter to convert our y axis to be in millions.


```python
from matplotlib.ticker import FuncFormatter

def milions(y, position):
    #Position does not matter for what we are doing here, but in other cases may be needed
    #Apply int
    s = str(y/1000000)
    return s
formatter = FuncFormatter(milions)
print(formatter)
```

```python
def plotCompoundingInterestWithWithdraw(P, rates, labels, withdraw):
    data = compoundWithdraw(P, rates, withdraw)
    for A, label in zip(data, labels):
        plt.plot(list(range(len(A))), A, label=label)        
    plt.xlabel('t')
    plt.ylabel('A (Bilion)')
    plt.legend()
    plt.gca().yaxis.set_major_formatter(formatter)
plotCompoundingInterestWithWithdraw(P,rates,["r=.02","r=.04","r=.06"],withdraw)
```

### Exchanging Compounding Interest at other frequency (Effective Return Rate)

Let's discuss a concept important to rates of return. When we are given a rate we generally assume **that it is annualized**, meaning that the rate we refer to is exactly what we will be getting at the end of a year. We can also, however, have rates that are **semi-annual, quarterly, monthly**, etc. What this means is that we compound at each of those periods.

###### Example

Let's take r = 10%, if is annualized, then we get 10% at the end of each year return. If it were semi-annually payments, we would get 5% halfway through the year and 5% at the end of the year. Multiplying the two periods of compounding, we would have (1.05)*(1.05) = 10.25%. We get a larger return because there is more compounding. If we have quarterly compounding, it becomes 2.5% every quarter which then becomes annualized to (1.025)*(1.025)*(1.025)*(1.025) = 10.38%. The generalized equation for annualizing a rate is as follows. We call the interest rate listed without considering compounding the nominal interest rate and we call the interest rate using compounding the effective interest rate.

![Exchange Compounding](2.png)

```python
def annualize(r, periods):
    periodRate = r / periods
    return ((1+periodRate)**periods) - 1

for period in [1, 4, 6, 12]:
    print(annualize(0.1, period))
```

Như vậy, với một APR nhất định, tần suất tính lãi càng nhiều thì EAR càng cao, nhưng cũng sẽ đụng tới một mức trần nào đó.  Nếu lãi được tính liên tục không ngừng nghỉ, EAR = eAPR – 1, trong đó e là hằng số (= 2,7182818…).  Với APR = 13,8%, lãi suất thực tế tối đa sẽ là EAR = 2,71828180,138 – 1 = 14,7976%, không thể nào “lên tới 15%-16%/năm”
